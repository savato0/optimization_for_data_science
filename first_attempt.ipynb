# Have fun

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulazione di errori per diversi tassi di convergenza
iterations = np.arange(1, 50)
linear_error = 0.8**iterations  # Lineare (retta in log-scale)
sublinear_error = 1/iterations  # Sublineare (curva lenta)

plt.semilogy(iterations, linear_error, label='Lineare (Q definita positiva)')
plt.semilogy(iterations, sublinear_error, label='Sublineare (Q semidefinita)')
plt.xlabel('Iterazione')
plt.ylabel('Errore f(x) - f*')
plt.legend()
plt.show()
```

# "See" the curvature

```python
import numpy as np

# Matrice Q del problema
Q = np.array([[10.0, 0.0], 
              [0.0,  1.0]]) # Molto più ripida su x che su y

# Calcolo autovalori
eigenvalues = np.linalg.eigvals(Q)
L = max(eigenvalues)
sigma = min(eigenvalues)

print(f"Curvatura massima (L): {L}")
print(f"Curvatura minima (sigma): {sigma}")
print(f"Rapporto di 'schiacciamento': {L/sigma}")
```
